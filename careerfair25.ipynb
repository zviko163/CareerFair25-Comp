```python
# input size, number of lines
N = [100, 200, 400, 800, 1000, 2000, 4000, 6000, 8000, 10000]
```

```python
# function of time taken by the program
def T(n, c, w):
    return 25 + 10*n + 3*w*n + c*n + n*n
```

```python
# plotting the graph
import matplotlib.pyplot as plt

# variations of words from 4000 to 100
t1 = [T(n, 100, 4000) for n in N]
t2 = [T(n, 100, 1000) for n in N]
t3 = [T(n, 100, 100) for n in N]

# variations of characters from 4000 to 100
t4 = [T(n, 4000, 100) for n in N]
t5 = [T(n, 1000, 100) for n in N]
# t6 = [T(n, 100, 100) for n in N]

# plotting the graph
plt.plot(N, t1, label='chars=100, words=4000')
plt.plot(N, t2, label='chars=100, words=1000')
plt.plot(N, t3, label='chars=100, words=100')

plt.plot(N, t4, label='chars=4000, words=100')
plt.plot(N, t5, label='chars=1000, words=100')
# plt.plot(N, t6, label='chars=100, words=100')


# plt.xticks(N)
plt.title('Runtime of the program')
plt.xlabel('Input File Size (no. of lines)')
plt.ylabel('Time taken (units)')
plt.legend()
plt.show()
```
